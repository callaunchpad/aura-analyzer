```python
import requests
import matplotlib.pyplot as plt
from PIL import Image

base_url = 'http://127.0.0.1:8000/'
```

```python
# Test outfit generation for all seasons

fig, axes = plt.subplots(4, 5)
axes = axes.flatten()
department = "menswear"

for i, color_season in enumerate(["spring", "summer", "autumn", "winter"]):
    body = f'''
    {{
    "Department": "{department}",
    "ColorSeason": "{color_season}",
    "n": 1
    }}
    '''

    response = requests.post(f'{base_url}/generate_outfit', body)

    items = response.json()
    # print(items)

    # Display each image
    for j, item in enumerate(items):
        item = item[0]         
        path = f'../color_analysis/grailed-dataset/images/{item['Id']}.jpg'
        image = Image.open(path)
        axes[i*5+j].imshow(image)
        axes[i*5+j].axis("off")

fig.set_figheight(10)
fig.set_figwidth(10)

plt.show()
```
