```python
import requests

base_url = 'http://127.0.0.1:8000/'
body = '''
{
  "Department": "womenswear",
  "ColorSeason": "spring",
  "num_outfits": 1
}
'''


response = requests.post(f'{base_url}/generate_outfit', body)
print(response.json())
```

```python
import matplotlib.pyplot as plt
from PIL import Image

# Create a figure and subplots
items_list = [x for x in response.json() if x != []]

fig, axes = plt.subplots(1, len(items_list))

# Display each image
for idx, items in enumerate(items_list):
    if len(items) > 0:
        item = items[0]
        # print(item)
        path = f'../color_analysis/grailed-dataset/images/{item['Id']}.jpg'
        image = Image.open(path)
        axes[idx].imshow(image)
        axes[idx].axis("off")

plt.show()
```
